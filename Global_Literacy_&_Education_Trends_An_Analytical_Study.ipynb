```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dataset Collection for adult


df_adult=pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_adult

```

```python
df_adult=df_adult.rename(columns = {'entity':'Country','year':'Year','adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99':'literate'})
df_adult
```

```python
#Dataset Collection for youth
df_youth=pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_youth
```

```python
df_youth=df_youth.rename(columns = {'entity':'Country','year':'Year','owid_region':'Region','youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m':'male_literate','youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f':'female_literate'})
df_youth
```

```python
df_youth['youth_literacy_avg'] = (
    df_youth['male_literate'] + df_youth['female_literate']
) / 2
df_youth
```

```python
#Dataset Collection for illiterate
df_Illiterate=pd.read_csv("https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_Illiterate
```

```python
df_Illiterate=df_Illiterate.rename(columns = {'entity':'Country','year':'Year','illiteracy_rate':'illiterate','literacy_rate':'literate'})
df_Illiterate
```

```python
#Dataset Collection for GDP
df_GDP=pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_GDP
```

```python
df_GDP=df_GDP.rename(columns = {'entity':'Country','year':'Year','owid_region':'Region','ny_gdp_pcap_pp_kd':'gdp_per_capita'})
df_GDP
```

```python
#Dataset Collection for AVOS(Average years of schooling)
df_AYOS=pd.read_csv("https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_AYOS
```

```python
df_AYOS=df_AYOS.rename(columns = {'entity':'Country','year':'Year','owid_region':'Region','mf_youth_and_adults__15_64_years__average_years_of_education':'avg_years_schooling'})
df_AYOS
```

```python
#since df_adult and df_illeterate is not having region columns, we are adding the region column by exracting and merging method:

df_region = df_youth[['Country', 'Region']].dropna().drop_duplicates()
df_region
```

```python
df_Illiterate = df_Illiterate.merge(
    df_region,
    on='Country',
    how='left'
)
df_Illiterate
```

# **Feature Engineering**

```python
#Illiteracy %
#to avoid the warning message
df_literacy = df_literacy.copy()

df_literacy['illiteracy_percent'] = (
    100 - df_literacy['adult_literacy_rate']
)

df_literacy
```

```python
#Literacy Gender Gap
df_literacy.loc[:,'literacy_gender_gap'] = (
    df_literacy['youth_male_literacy_rate'] -
    df_literacy['youth_female_literacy_rate']
)
df_literacy
```

```python
#GDP per Schooling Year
df_gdp_schooling['gdp_per_schooling_year'] = (
    df_gdp_schooling['gdp_per_capita_usd'] /
    df_gdp_schooling['average_years_of_schooling']
)
df_gdp_schooling
```

```python
#Education Index
df_merged = pd.merge(
    df_literacy,
    df_gdp_schooling,
    on=['Country', 'Year','Region'],
    how='inner'
)

df_merged['education_index'] = (
    (df_merged['adult_literacy_rate'] / 100) *
    (df_merged['average_years_of_schooling'] /
     df_merged['average_years_of_schooling'].max())
)
df_merged
```

```python
#Youth Literacy Average
df_literacy.loc[:,'youth_literacy_average'] = (
    df_literacy['youth_male_literacy_rate'] +
    df_literacy['youth_female_literacy_rate']
) / 2
df_literacy
```

```python
#Literacy Growth Rate (Year-over-Year)
df_literacy = df_literacy.sort_values(['Country', 'Year'])

df_literacy['literacy_growth_rate'] = (
    df_literacy.groupby('Country')['adult_literacy_rate']
    .pct_change() * 100
)
df_literacy
```

```python
#Literacy to GDP Ratio
df_merged['literacy_gdp_ratio'] = (
    df_merged['adult_literacy_rate'] /
    df_merged['gdp_per_capita_usd']
)
df_merged
```

```python
#Education Efficiency Score
df_merged['education_efficiency'] = (
    df_merged['gdp_per_capita_usd'] *
    df_merged['adult_literacy_rate']
) / 100
df_merged
```

#Data Storage in SQL

```python
import sqlite3

# Connect once
conn = sqlite3.connect("education_analysis.db")
cursor = conn.cursor()

# Create all tables
cursor.executescript("""
CREATE TABLE IF NOT EXISTS literacy_rates (
    country TEXT NOT NULL,
    year INTEGER NOT NULL,
    adult_literacy_rate REAL,
    youth_male_literacy_rate REAL,
    youth_female_literacy_rate REAL,
    youth_literacy_average REAL,
    literacy_gender_gap REAL,
    literacy_growth_rate REAL,
    PRIMARY KEY (country, year)
);

CREATE TABLE IF NOT EXISTS illiteracy_population (
    country TEXT NOT NULL,
    year INTEGER NOT NULL,
    total_literacy_rate REAL,
    illiteracy_rate REAL,
    PRIMARY KEY (country, year)
);

CREATE TABLE IF NOT EXISTS gdp_schooling (
    country TEXT NOT NULL,
    year INTEGER NOT NULL,
    gdp_per_capita_usd REAL,
    average_years_of_schooling REAL,
    gdp_per_schooling_year REAL,
    PRIMARY KEY (country, year)
);
""")

# Commit changes and close connection
conn.commit()
```

```python
#1️ Top 5 countries with highest adult literacy in 2020
import sqlite3
import pandas as pd

conn = sqlite3.connect("education_analysis.db")

df = pd.read_sql("""
SELECT country, adult_literacy_rate
FROM literacy_rates
WHERE Year = 2020
ORDER BY adult_literacy_rate DESC
LIMIT 5;
""", conn)


df
```

```python
#Countries where female youth literacy < 80%
df = pd.read_sql("""
SELECT country, year, youth_female_literacy_rate
FROM literacy_rates
WHERE youth_female_literacy_rate < 80
ORDER BY youth_female_literacy_rate ASC;
""", conn)


df
```

```python
#Average adult literacy per continent (OWID region)

df = pd.read_sql("""

SELECT
    Region AS continent,
    AVG(adult_literacy_rate) AS avg_adult_literacy
FROM
    literacy_rates
GROUP BY
    Region
ORDER BY
    avg_adult_literacy DESC;

""", conn)


df
```

```python
#4. Countries with illiteracy % > 20% in 2000.
df = pd.read_sql("""
SELECT
    country,
    illiteracy_rate
FROM
    illiteracy_population
WHERE
    year = 2000
    AND illiteracy_rate > 20
ORDER BY
    illiteracy_rate DESC;

""", conn)


df
```

```python
#5. Trend of illiteracy % for India (2000–2020).

df = pd.read_sql("""
SELECT
    year,
    illiteracy_rate
FROM
    illiteracy_population
WHERE
    country = 'India'
    AND year BETWEEN 2000 AND 2020
ORDER BY
    year;

""", conn)


df
```

```python
#6. Top 10 countries with largest illiterate population in the last year.
df = pd.read_sql("""
SELECT
    country,
    illiteracy_rate
FROM
    illiteracy_population
WHERE
    year = (SELECT MAX(year) FROM illiteracy_population)
ORDER BY
    illiteracy_rate DESC
LIMIT 10;

""", conn)


df
```

#gdp_schooling


```python
#7. Find countries with avg_years_schooling > 7 and gdp_per_capita < 5000.
df = pd.read_sql("""
SELECT
    country,
    year,
    average_years_of_schooling,
    gdp_per_capita_usd
FROM
    gdp_schooling
WHERE
    average_years_of_schooling > 7
    AND gdp_per_capita_usd < 5000
ORDER BY
    country;

""", conn)


df
```

```python
# 8. Rank countries by GDP per schooling for the year 2020.
df = pd.read_sql("""
SELECT
    country,
    gdp_per_schooling_year
FROM
    gdp_schooling
WHERE
    year = 2020
ORDER BY
    gdp_per_schooling_year DESC;

""", conn)


df
```

```python
#9. Find global average schooling years per year.

df = pd.read_sql("""
SELECT
    year,
    AVG(average_years_of_schooling) AS global_avg_schooling
FROM
    gdp_schooling
GROUP BY
    year
ORDER BY
    year;

""", conn)


df
```

#Join Queries


```python
#10.Top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling (<6)
df = pd.read_sql("""
SELECT
    g.country,
    g.gdp_per_capita_usd,
    g.average_years_of_schooling
FROM
    gdp_schooling g
WHERE
    g.year = 2020
    AND g.average_years_of_schooling < 6
ORDER BY
    g.gdp_per_capita_usd DESC
LIMIT 10;

""", conn)


df
```

```python
#Countries where the illiterate population is high despite having more than 10 average years of schooling
df = pd.read_sql("""
SELECT
    i.country,
    i.year,
    i.illiteracy_rate,
    g.average_years_of_schooling
FROM
    illiteracy_population i
JOIN
    gdp_schooling g
ON
    i.country = g.country AND i.year = g.year
WHERE
    g.average_years_of_schooling > 10
    AND i.illiteracy_rate > 20  -- assuming >20% is "high"
ORDER BY
    i.illiteracy_rate DESC;

""", conn)

#there is no result because MAX(average_years_of_schooling) is 13.74 and MIN(illiteracy_rate) is 89.0

df
```

```python
#Compare literacy rates and GDP per capita growth for a selected country over the last 20 years
df = pd.read_sql("""
SELECT
    l.year,
    l.adult_literacy_rate,
    g.gdp_per_capita_usd
FROM
    literacy_rates l
JOIN
    gdp_schooling g
ON
    l.country = g.country AND l.year = g.year
WHERE
    l.country = 'India'
    AND l.year >= (SELECT MAX(year) FROM literacy_rates) - 20
ORDER BY
    l.year;

""", conn)


df
```

```python
#Difference between youth literacy male and female rates for countries with GDP per capita > $30,000 in 2020
df = pd.read_sql("""
SELECT
    l.country,
    l.year,
    l.youth_male_literacy_rate,
    l.youth_female_literacy_rate,
    (l.youth_male_literacy_rate - l.youth_female_literacy_rate) AS gender_gap
FROM
    literacy_rates l
JOIN
    gdp_schooling g
ON
    l.country = g.country AND l.year = g.year
WHERE
    l.year = 2020
    AND g.gdp_per_capita_usd > 30000
ORDER BY
    gender_gap DESC;

""", conn)


df
```
